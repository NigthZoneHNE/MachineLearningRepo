```python
import numpy as np
```

Matrix Multiplication

```python
a = np.array([[1,2,3],
              [1,2,3]])

b = np.array([[1,2],
              [1,2],
              [1,2]])


print(f'array{a @ b}')

#or :

np.dot(a,b)
```

Matrix Summation

```python
# summed = a + b --> should have same shapes

# print(summed)

total = np.sum(a) #-> All elements of a
print(total)

col_sum = np.sum(a,axis=0) # sum of columns
print(col_sum)

row_sum = np.sum(a,axis=1) # sum of rows
print(row_sum)
```

Matrix Transpose

```python
A_T = a . T # -> .T is transposer
A_T
```

Identity Matrix

```python
I = np.eye(2)
I
```

Inverse Matrix

```python
S = np.array([[1,2],[3,4]])
S_inv = np.linalg.inv(S) # -> matrix should be sqare and has 0 in result of determinant
print(S)
S_inv
```

Symmetric Functions

```python
is_symmetric = np.allclose(S, S.T)
is_symmetric
```

Scaling A Matrix

```python
λ = 3
C = S * λ
C
```

Matrix Concatenation

```python
x = np.array([[1,2],[3,4]])
y = np.array([[5,6],[7,8]])

j = np.concatenate((x,y),axis=1)
l = np.concatenate((x,y),axis=0)
print(j)
print(l)
```
